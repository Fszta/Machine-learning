```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Chargement et affichage du dataset

```python
data_diabetes = datasets.load_breast_cancer()
X_diabetes = data_diabetes.data
y_diabetes = data_diabetes.target

print("Nombre de features : %s" %len(data_diabetes.feature_names))
print(data_diabetes.feature_names)

print("Nombre de classes: %s" %len(data_diabetes.target_names))
print(data_diabetes.target_names)
```

## Plot des deux clusters selon les 3 premiers features

```python
k_means = KMeans(n_clusters=2)
k_means.fit(X_diabetes)
y_kmeans = k_means.predict(X_diabetes)
X = X_diabetes

fig = plt.figure(1, figsize=(10, 10))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
plt.title("Dataset cancer")
labels = k_means.labels_

ax.scatter(X[:,0],X[:,1],X[:,2], c = y_kmeans, edgecolors="k", s = 60)
```
