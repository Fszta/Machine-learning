```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import classification_report
from sklearn.datasets import load_wine
from xgboost import XGBClassifier
```

```python
data_wine = load_wine()
feature_names = data_wine['feature_names']
```

```python
# Convert data into pandas dataframe
data_wine = pd.DataFrame(data = np.c_[data_wine['data'], data_wine['target']],
                         columns= data_wine['feature_names'] + ['target'])
```

```python
# Get features and labels
X = data_wine.drop('target',axis=1)
y = data_wine['target']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20)
```

```python
xgboost = XGBClassifier()

# Create param grid
param_grid = {
            'min_child_weight': [1, 5, 10,20],
            'gamma': [0.5, 1, 1.5, 2, 5],
            'subsample': [0.6, 0.8, 1.0],
            'colsample_bytree': [0.6, 0.8, 1.0],
            'max_depth': [3, 4, 5, 6, 7]
            }

random_search = RandomizedSearchCV(xgboost, param_distributions=param_grid, 
                                   n_iter=10, n_jobs=4, 
                                   cv=3, verbose=2, random_state=42)

best_model = random_search.fit(X_train,y_train)

print('Best score :',best_model.best_score_,'\n')

print('Best estimator:\n',best_model.best_estimator_,'\n')
```

```python
y_pred = best_model.predict(X_test)
print('Classification_report : \n',classification_report(y_test, y_pred))
```

```python
plt.title('Features importances')
df = pd.DataFrame(best_model.best_estimator_.feature_importances_,X.columns).reset_index()
df = df.rename(columns={'index':'features', 0:'coef'})
df = df.sort_values(by='coef', ascending=True)
sns.barplot(x='coef', y='features', data=df, palette='pastel')
```
